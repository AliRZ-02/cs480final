# Curve: Training & Validation Loss compared to Epoch

```python
import matplotlib.pyplot as plt
import json
```

```python
original_model_losses = {}
with open('overfit.json', 'r') as f:
    original_model_losses = json.load(f)

epochs = range(1, 11)
plt.plot(epochs, original_model_losses['train'], 'bo-', label='Training Loss')
plt.plot(epochs, original_model_losses['val'], 'r^-', label='Validation Loss')

plt.title('Original Model: Train and Val Loss Per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
final_model_losses = {}
with open('overfitV7.json', 'r') as f:
    final_model_losses = json.load(f)

epochs = range(1, 11)

f, ax = plt.subplots(2, sharex=True)

ax[1].plot(epochs, final_model_losses['train'][:10], 'bo-', label='Training Loss')
ax[1].plot(epochs, final_model_losses['val'][:10], 'r^-', label='Validation Loss')
ax[1].set_title('Final Model: Train and Val Loss In first 10 Epochs')
ax[0].plot(epochs, original_model_losses['train'], 'bo-', label='Training Loss')
ax[0].plot(epochs, original_model_losses['val'], 'r^-', label='Validation Loss')
ax[0].set_title('Original Model: Train and Val Loss Per Epoch')


plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from collections import defaultdict

final_model_losses = defaultdict(lambda: list())
with open('overfitV7.json', 'r') as f:
    data = json.load(f)

    final_model_losses["train"].extend(data["train"])
    final_model_losses["val"].extend(data["val"])

with open('overfitV9.json', 'r') as f:
    data = json.load(f)

    final_model_losses["train"].extend(data["train"])
    final_model_losses["val"].extend(data["val"])

with open('overfitV10.json', 'r') as f:
    data = json.load(f)

    final_model_losses["train"].extend(data["train"])
    final_model_losses["val"].extend(data["val"])

print(min([(val, i) for i, val in enumerate(final_model_losses["val"])]))

epochs = range(8, 161)
plt.plot(epochs, final_model_losses['train'][7:], 'bo-', label='Training Loss')
plt.plot(epochs, final_model_losses['val'][7:], 'r^-', label='Validation Loss')

plt.title('Final Model: Train and Val Loss Per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
